**to replace the null values we replace it either with mean or median depends on the type**
generally mean can get effected by outliers but median don't

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

**we have upper and lower cases and spaces inbetween the words and this is ineffecient** 
we should make it all low or upper cases and replace the spaces with underscore

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)':'purphase_amount'})
```

```python
df.columns
```

**qcut :splits ages into the no you want based on the data distribution and asigns the label we defined**

create column age_group
useful to understand customer behavior and marketing

```python
labels = ['Young Adult','Adult','Middle_aged','Senior']
df['age_group']= pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(11)
```
